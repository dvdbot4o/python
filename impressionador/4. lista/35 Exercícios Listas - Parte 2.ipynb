# Exercícios

## 1. Mudança de Carga Tributária

- Reformas e mudanças de cargas tributárias são bem comuns no Brasil.

Digamos que você trabalhe em uma empresa de ecommerce

No Brasil, o imposto sobre livros é zerado. De um ano para o outro, o governo criou um novo imposto que incide em 10% sobre o valor dos livros e agora você precisa alterar o registro dos preços dos livros da empresa para garantir que esse imposto vai ser repassado para o preço final do produto.

Crie um código que recalcule o valor do livro da sua lista de produtos e ajuste na tabela.

Além disso, calcule qual vai ser o impacto financeiro da criação desse imposto para a empresa (ou seja, quanto que o imposto vai aumentar de custo para a empresa)

Obs: para facilitar, colocamos apenas 1 livro na lista, mas em breve vamos aprender um for que vai adaptar esse cenário para qualquer quantidade de livros na sua lista.

Obs2: Seu código deve funcionar mesmo que não haja livros na lista de produtos da empresa

```python
# Lista com os nomes dos produtos disponíveis no e-commerce
produtos = ['computador', 'livro', 'tablet', 'celular', 'tv', 'ar condicionado', 'alexa', 'máquina de café', 'kindle']

# Lista de listas, onde cada sublista contém:
# [quantidade de vendas no mês, preço do produto]
produtos_ecommerce = [
    [10000, 2500],   # computador: 10.000 vendas, preço 2500
    [50000, 40],     # livro: 50.000 vendas, preço 40
    [7000, 1200],    # tablet: 7.000 vendas, preço 1200
    [20000, 1500],   # celular: 20.000 vendas, preço 1500
    [5800, 1300],    # tv: 5.800 vendas, preço 1300
    [7200, 2500],    # ar condicionado: 7.200 vendas, preço 2500
    [200, 800],      # alexa: 200 vendas, preço 800
    [3300, 700],     # máquina de café: 3.300 vendas, preço 700
    [1900, 400]      # kindle: 1.900 vendas, preço 400
]

# Verifica se o produto "livro" está presente na lista de produtos
if "livro" in produtos:
    # Obtém o índice do produto "livro" na lista 'produtos'
    i_livros = produtos.index("livro")

    # Usa o índice para acessar os dados do livro na lista 'produtos_ecommerce'
    # 'livro' será uma lista [quantidade_vendas, preço]
    livro = produtos_ecommerce[i_livros]

    # Exibe uma mensagem formatada:
    # - Calcula a quantidade de vendas com um acréscimo de 10% (livro[0] + livro[0]*0.10)
    # - Mostra o preço do livro (livro[1])
    print(f" o livro teve cerca de {livro[0] + livro[0]*0.10:.2f} com cerca de é {livro[1]} copias vendidas ")

```
