Digamos que você está construindo o controle de produtos da Apple.<br>
E a Apple lançou o IPhone 11 e irá tirar dos seus estoques o IPhone X

```python
produtos = ['apple tv', 'mac', 'iphone x', 'IPad', 'apple watch', 'mac book', 'airpods']

print(produtos)

# acrescentar

produtos.append("Iphone 11")

print(produtos)

# remover remove

produtos.remove("Iphone 11")

# remover pop

print(produtos)

produtos.pop(-1)

print(produtos)
```

## Existem 2 formas de tratar o erro:

1. Criar um if para evitar que ele aconteça

2. Esperar que ele possa acontecer e tratar caso o erro aconteça com:

```python
def dividir(a, b):
    try:
        resultado = a / b
    except ZeroDivisionError:
        return "Erro: divisão por zero não permitida"
    except TypeError:
        return "Erro: ambos os valores devem ser números"
    else:
        return f"Resultado: {resultado}"
    finally:
        print("Função dividir executada")

# Testes
print(dividir(10, 2))   # Resultado: 5.0
print(dividir(10, 0))   # Erro: divisão por zero
print(dividir(10, "a")) # Erro: tipo inválido
```
