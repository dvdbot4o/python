```python
"""Exercício 01
Faça um programa que peça dois números e imprima o maior deles."""

n1 = float(input(" digite um numero 1 "))

n2 = float(input(" digite um numero 2 "))

if n1 > n2:
  print(f" o numero {n1} é maior que o {n2}")
elif n1 < n2:
  print(f" o numero {n2} é maior que o {n1}")
else:
  print("os numeros sao iguais ")
```

```python
"""Exercício 02
Faça um programa que peça um valor e mostre na tela se o valor é positivo ou negativo."""

n = float(input(" digite um numero : "))

if n >= 0:
  print(" esse numero é positivo ")
else:
  print(" esse numero é negativo ")
```

```python
"""Exercício 03
Faça um programa que verifique se uma letra digitada é "F" ou "M". Conforme a letra escrever:

F - Feminino
M - Masculino
Sexo Inválido."""

sexo = input(" digite F para feminino ou M para masculino : ")

if sexo == "f" or sexo == "F":
  print(" você é do sexo feminino  ")
elif sexo == "m" or sexo == "M":
  print(" você é do sexo masculino ")
else:
  print(" digitou errado ")
```

```python
"""Exercício 06
Faça um programa que leia três números e mostre o maior deles:"""

numeros = []

for i in range(3):

    numeros.append(float(input(f" digite o numero {i + 1} ° ")))

print(numeros)

menor = numeros[0]

for i in numeros:
    if i < menor:
        menor = i
print(f" o menor numero da lista  é {menor}")
```

```python
"""Exercício 07
Faça um programa que leia três números e mostre o maior e o menor deles:"""

numeros = []

for i in range(3):

    numeros.append(float(input(f" digite o numero {i + 1} ° ")))

print(numeros)

menor = numeros[0]
maior = 0
for i in numeros:
    if i < menor:
        menor = i
    if i > maior:
        maior = i
print(f" o menor numero da lista  é {menor} e o maior é {maior}")

```
