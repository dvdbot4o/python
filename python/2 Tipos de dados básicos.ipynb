```python
# --- Inteiros (int) ---
# Comentário explicando que inteiros representam números sem parte decimal.

numero_inteiro = 42  # Define número inteiro
print(numero_inteiro)           # Imprime: 42
print(type(numero_inteiro))     # Imprime tipo: <class 'int'>

# Operações comuns com inteiros:
print(numero_inteiro + 8)       # Soma: 50
print(numero_inteiro - 10)      # Subtração: 32
print(numero_inteiro * 2)       # Multiplicação: 84
print(numero_inteiro // 5)      # Divisão inteira: 8
print(numero_inteiro % 5)       # Resto da divisão: 2
print(numero_inteiro ** 2)      # Potência: 1764

# --- Float (ponto flutuante) ---
# Comentário explicando que floats representam números reais com parte decimal.

numero_float = 3.14159  # Define número float
print(numero_float)             # Imprime: 3.14159
print(type(numero_float))       # Imprime tipo: <class 'float'>

# Operações com floats:
print(numero_float + 2.0)       # Soma: 5.14159
print(numero_float * 2)         # Multiplicação: 6.28318
print(numero_float / 2)         # Divisão: 1.570795

# Atenção: operações com floats podem ter pequenas imprecisões
print(0.1 + 0.2)                # Imprime: 0.30000000000000004

# --- Strings (str) ---
# Comentário explicando que strings são sequências de caracteres para texto.

texto = "Olá, mundo!"  # Define string
print(texto)                   # Imprime: Olá, mundo!
print(type(texto))             # Imprime tipo: <class 'str'>

# Strings podem usar aspas simples ou duplas:
str1 = 'Python'
str2 = "Python"
print(str1 == str2)            # True (iguais)

# Caracteres especiais com escape:
print("Linha 1\nLinha 2")      # Quebra de linha
print("Aspas: \"Python\"")    # Aspas dentro da string

# Operações comuns com strings:
print(texto + " Vamos aprender.")  # Concatenação
print(texto * 3)                   # Repetição

# Indexação e fatiamento:
print(texto[0])                   # Primeiro caractere: 'O'
print(texto[-1])                  # Último caractere: '!'
print(texto[0:4])                 # Fatiamento: 'Olá,'
print(texto[5:])                  # Fatiamento do índice 5 até o fim: 'mundo!'

# Métodos úteis de string:
print(texto.lower())              # Tudo minúsculo
print(texto.upper())              # Tudo maiúsculo
print(texto.replace("mundo", "Python"))  # Substituição

# --- Booleanos (bool) ---
# Comentário explicando que booleanos representam True ou False.

verdadeiro = True
falso = False
print(verdadeiro)                # True
print(type(verdadeiro))          # <class 'bool'>

# Booleanos em condições:
print(10 > 5)                   # True
print(10 == 5)                  # False

# Operadores lógicos:
print(True and False)           # False
print(True or False)            # True
print(not True)                 # False

# Conversão entre tipos:
print(bool(0))                  # False (zero é falso)
print(bool(1))                  # True  (não zero é verdadeiro)
print(bool(""))                 # False (string vazia é falsa)
print(bool("texto"))            # True  (string não vazia é verdadeira)

# --- Conversão entre tipos básicos ---

# De string para int/float:
num_str = "123"
num_int = int(num_str)
num_float = float(num_str)
print(num_int)                  # 123
print(type(num_int))            # <class 'int'>
print(num_float)                # 123.0
print(type(num_float))          # <class 'float'>

# De int/float para string:
num = 456
num_str = str(num)
print(num_str)                 # '456'
print(type(num_str))           # <class 'str'>

# De booleano para int:
print(int(True))               # 1
print(int(False))              # 0

# De int para booleano:
print(bool(0))                # False
print(bool(42))               # True

# --- Dicas importantes ---

# 1. Python é fortemente tipado: não permite operações inválidas entre tipos diferentes sem conversão explícita.
# Exemplo que gera erro:
# print("Idade: " + 30)  # TypeError

# Correto:
print("Idade: " + str(30))   # Idade: 30

# 2. Use type() para verificar o tipo de qualquer variável.

# 3. Strings são imutáveis: não é possível alterar um caractere diretamente.
s = "Python"
# s[0] = "J"  # Erro! TypeError
# Para modificar, crie nova string:
s_modificada = "J" + s[1:]
print(s_modificada)           # Jython

# 4. Operações entre int e float resultam em float:
print(type(5 + 2.0))          # <class 'float'>

# 5. Booleanos são subclasses de inteiros:
print(True + True)            # 2 (True equivale a 1)

# --- Resumo rápido dos tipos básicos ---
# int: números inteiros
# float: números reais com ponto decimal
# str: texto
# bool: valores lógicos True/False

# Entender esses tipos é fundamental para manipular dados corretamente em Python.
```
