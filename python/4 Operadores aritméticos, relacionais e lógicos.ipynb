```python
# --- Operadores Aritméticos ---
# Comentário explicando que são usados para realizar operações matemáticas básicas.

a = 15  # Define variável a com valor 15
b = 4   # Define variável b com valor 4

print(a + b)   # Imprime soma: 19  (adição)
print(a - b)   # Imprime subtração: 11  (subtração)
print(a * b)   # Imprime multiplicação: 60  (multiplicação)
print(a / b)   # Imprime divisão real: 3.75 (divisão real, sempre retorna float)
print(a // b)  # Imprime divisão inteira: 3   (descarta parte decimal)
print(a % b)   # Imprime módulo: 3   (resto da divisão)
print(a ** b)  # Imprime exponenciação: 50625 (15 elevado a 4)

# Observação: divisão (/) sempre retorna float, mesmo se a divisão for exata
print(10 / 2)  # Imprime 5.0 (float)

# --- Operadores Relacionais (Comparação) ---
# Comentário explicando que retornam valores booleanos (True ou False) comparando dois valores.

x = 10  # Define variável x
y = 20  # Define variável y

print(x == y)   # False (igualdade)
print(x != y)   # True  (diferença)
print(x > y)    # False (maior que)
print(x < y)    # True  (menor que)
print(x >= 10)  # True  (maior ou igual)
print(y <= 20)  # True  (menor ou igual)

# Comparações podem ser encadeadas:
print(5 < x < 15)  # True (x está entre 5 e 15)

# --- Operadores Lógicos ---
# Comentário explicando que são usados para combinar expressões booleanas.

p = True   # Define variável booleana p
q = False  # Define variável booleana q

print(p and q)   # False (E lógico: True somente se ambos forem True)
print(p or q)    # True  (OU lógico: True se pelo menos um for True)
print(not p)     # False (negação)

# Exemplos com expressões:
idade = 25  # Define idade
tem_carteira = True  # Define se tem carteira

# Verifica se a pessoa pode dirigir (idade >= 18 e tem carteira)
pode_dirigir = (idade >= 18) and tem_carteira
print(pode_dirigir)  # True

# --- Precedência dos operadores ---
# Comentário explicando a ordem em que os operadores são avaliados:
# 1. Parênteses ()
# 2. Exponenciação **
# 3. Multiplicação *, Divisão /, Divisão inteira //, Módulo %
# 4. Adição +, Subtração -
# 5. Operadores relacionais
# 6. Operadores lógicos: not, and, or

# Exemplo:
resultado = 3 + 4 * 2  # Multiplicação antes da adição
print(resultado)       # 11

resultado = (3 + 4) * 2  # Parênteses alteram a ordem
print(resultado)         # 14

# --- Operadores de atribuição compostos ---
# Comentário explicando que atualizam o valor da variável com uma operação.

contador = 10  # Inicializa contador com 10
contador += 5   # contador = contador + 5
print(contador)  # Imprime 15

contador *= 2   # contador = contador * 2
print(contador)  # Imprime 30

contador %= 7   # contador = contador % 7
print(contador)  # Imprime 2

# --- Operadores de identidade e associação ---
# Comentário explicando:
# identity (is, is not): verificam se duas variáveis apontam para o mesmo objeto
# membership (in, not in): verificam se um valor está contido em uma coleção

x = [1, 2, 3]  # Lista x
y = x          # y referencia o mesmo objeto que x
z = [1, 2, 3]  # Nova lista com mesmo conteúdo

print(x is y)       # True (mesmo objeto)
print(x is z)       # False (objetos diferentes, mesmo conteúdo)
print(x == z)       # True (conteúdo igual)

print(2 in x)       # True (2 está na lista x)
print(5 not in x)   # True (5 não está na lista x)

# --- Exemplos práticos combinando operadores ---

# Verificar se um número é par e positivo
numero = 8  # Define número
eh_par = (numero % 2 == 0)  # Verifica se é par
eh_positivo = (numero > 0)  # Verifica se é positivo
print(eh_par and eh_positivo)  # True

# Verificar se uma letra está em uma palavra
palavra = "Python"  # Define palavra
letra = "y"        # Define letra
print(letra in palavra)        # True

# Negar uma condição
chovendo = False  # Define variável chovendo
nao_chovendo = not chovendo  # Negação
print(nao_chovendo)            # True

# --- Resumo ---
# Comentários resumindo conceitos:
# - Operadores aritméticos manipulam números
# - Operadores relacionais comparam valores e retornam booleanos
# - Operadores lógicos combinam expressões booleanas
# - Precedência define a ordem de avaliação
# - Operadores compostos facilitam atualizações de variáveis
```
