```python
# --- Controle de fluxo com condicionais ---
# Comentário explicando que condicionais permitem que o programa tome decisões e execute blocos diferentes
# dependendo se condições são True ou False.

# Estrutura básica:
# if condição:
#     bloco de código 1
# elif outra_condição:
#     bloco de código 2
# else:
#     bloco de código 3

# --- Exemplo simples ---
idade = 20  # Define variável idade com valor 20

if idade >= 18:  # Verifica se idade é maior ou igual a 18
    print("Maior de idade")  # Executa se condição for True
else:  # Executa se condição for False
    print("Menor de idade")  # Imprime mensagem alternativa

# Saída: Maior de idade

# --- Usando elif para múltiplas condições ---
nota = 75  # Define variável nota com valor 75

if nota >= 90:  # Se nota maior ou igual a 90
    print("Excelente")  # Imprime Excelente
elif nota >= 70:  # Se nota maior ou igual a 70
    print("Bom")  # Imprime Bom
elif nota >= 50:  # Se nota maior ou igual a 50
    print("Regular")  # Imprime Regular
else:  # Se nenhuma condição anterior for verdadeira
    print("Insuficiente")  # Imprime Insuficiente

# Saída: Bom

# --- Explicação ---
# Comentário explicando que Python avalia condições na ordem.
# Quando encontra a primeira condição True, executa o bloco e ignora o restante.

# --- Condições compostas com operadores lógicos ---
temperatura = 25  # Define temperatura
chuva = False  # Define variável chuva como False

if temperatura > 20 and not chuva:  # Verifica se temperatura > 20 e não está chovendo
    print("Dia agradável para passear")  # Imprime mensagem positiva
else:  # Caso contrário
    print("Melhor ficar em casa")  # Imprime mensagem alternativa

# Saída: Dia agradável para passear

# --- Condições aninhadas ---
numero = 15  # Define número

if numero > 0:  # Verifica se número é positivo
    if numero % 2 == 0:  # Verifica se número é par
        print("Número positivo e par")  # Imprime mensagem
    else:  # Se não for par
        print("Número positivo e ímpar")  # Imprime mensagem
else:  # Se número não for positivo
    print("Número não é positivo")  # Imprime mensagem

# Saída: Número positivo e ímpar

# --- Condicional em uma linha (expressão condicional ou ternária) ---
idade = 17  # Define idade

status = "Maior" if idade >= 18 else "Menor"  # Atribui "Maior" se idade >= 18, senão "Menor"
print(status)  # Imprime status: Menor

# --- Boas práticas ---
# Comentários com boas práticas:
# - Use indentação correta (4 espaços) para definir blocos
# - Evite muitas condições aninhadas; prefira funções para organizar o código
# - Use elif para múltiplas condições exclusivas

# --- Exemplo prático: verificar categoria de atleta ---
idade = 12  # Define idade do atleta

if idade < 10:  # Se idade menor que 10
    categoria = "Infantil"  # Categoria Infantil
elif idade < 15:  # Se idade menor que 15
    categoria = "Juvenil"  # Categoria Juvenil
elif idade < 20:  # Se idade menor que 20
    categoria = "Júnior"  # Categoria Júnior
else:  # Caso contrário
    categoria = "Sênior"  # Categoria Sênior

print(f"Categoria: {categoria}")  # Imprime categoria: Juvenil

# --- Erros comuns ---
# Comentários explicando erros comuns:
# Esquecer os dois pontos (:) após if, elif, else
# Exemplo errado:
# if idade >= 18
#     print("Maior de idade")  # SyntaxError

# Esquecer a indentação correta:
# if idade >= 18:
# print("Maior de idade")  # IndentationError

# --- Resumo ---
# Comentários resumindo conceitos:
# - if executa bloco se condição for True
# - elif verifica outras condições se anteriores forem False
# - else executa bloco padrão se nenhuma condição for True
# - operadores lógicos permitem combinar condições
# - indentação define blocos de código
```
