```python
produtos = ["iphone", "ipad", "airpod", "macbook"]
precos = [7000, 10000, 2500, 14000]
```

### For item in lista

```python
# preco com imposto

for i, produto in enumerate(produtos):
    imposto = precos[i] * 0.20 + precos[i] 
    print(f" o preço do produto {produto} com imposto é de R${imposto:.2f}")

```

### For i in range

```python
# preco de cada produto
for i, produto in enumerate(produtos):
    imposto = precos[i] * 0.20 + precos[i] 
    print(f" o preço do produto {produto} com imposto é de R${imposto:.2f}")
```
